#Series de Fourier

Sympy es una librería de matemática simbólica de python. Esta librería es capaz de resolver muchos tipos de problemas que van desde simples problemas álgebráicos y aritméticos hasta soluciones de ecuaciones diferenciales. Entre los usos que se le pueden dar a la librería se encuentran el cálculo de series de Fourier.

Posteriormente se importará la librería como se muestra a continuación.

````python
from sympy import fourier_series, pi, plot, Piecewise, exp, simplify
from sympy.abc import x
````

Se escribe la función de la que se desea obtener su desarrollo de la serie de Fourier. Para ello usamos la función *Piecewise()* para generar una función a trozos o función escalón.

#Ejemplo

````python
f=Piecewise(
    (0, x < -2),
    (x+2,(x >= -2) & (x < 0)),
    (0,(x<=2)))
````

````python
plot(f, (x,-3,2),line_color ='red')
````

Al usar la función *fourier_series(f (x,li,ls))* se genera la serie de Fourier de la función *f* con respecto a la variable *x* en el intervalo (li,ls). Posteriormente, el método *truncate* devuelve la serie de furier hasta el número de términos que se le indique.

````python
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
````

````python
s
````

````python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
````

````python
p.show()
````

En la gráfica anterior se observan 3 desarrollos de Fourier hasta un determinado término de la serie.

#Ejercicio $$ f(x) = e^{2x} $$

##Serie de Fourier Seno

Es necesario extender la función $f(x) = e^{2x}$ a una función por partes que sea impar.

````python
gs=Piecewise(
    (0, x < -1),
    (-exp(2*-x),(x >= -1) & (x < 0)),
    (exp(2*x),(x >= 0) & (x < 1)),
    (0,(x>=1)))

plot(gs, (x,-1.5,1.5),line_color ='red')
````

````python
s = fourier_series(gs, (x, -1, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
s4 = s.truncate(n = 10)
s5 = s.truncate(n = 20)
p = plot(gs, s1, s2, s3, (x, -pi, pi), show=False, legend=True)

s
````

````python
s.simplify()
````

````python
p[0].label = '$e^{2x}$'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'

p.show()
````

````python
p = plot(gs, s4, s5, (x, -pi, pi), show=False, legend=True)
p[0].label = '$e^{2x}'
p[1].label = 'n=10'
p[2].label = 'n=20'
p.show()
````

#Serie de Fourier Coseno


Es necesario extender la función $f(x) = e^{2x}$ a una función por partes que sea par.

````python
gc=Piecewise(
    (0, x < -1),
    (exp(2*-x),(x >= -1) & (x < 0)),
    (exp(2*x),(x >= 0) & (x < 1)),
    (0,(x>=1)))

plot(gc, (x,-1.5,1.5),line_color ='red')
````

````python
ss = fourier_series(gc, (x, -1, 1))
s1 = ss.truncate(n = 3)
s2 = ss.truncate(n = 5)
s3 = ss.truncate(n = 7)
p = plot(gc, s1, s2, s3, (x, -pi, pi), show=False, legend=True)

ss
````

````python
simplify(ss)
````

````python
p[0].label = '$e^{2x}'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'

p.show()
````

````python
s4 = ss.truncate(n = 10)
s5 = ss.truncate(n = 20)
p = plot(gc, s4, s5, (x, -pi, pi), show=False, legend=True)
p[0].label = '$e^{2x}'
p[1].label = 'n=10'
p[2].label = 'n=20'
p.show()
````
